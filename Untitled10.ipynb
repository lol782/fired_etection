```python
import cv2
from ultralytics import YOLO

# Load the YOLO model
model = YOLO("/content/fire/fire/runs/detect/train/weights/best.pt")  # Replace with your trained model path

# Open webcam (0 = default camera, 1 = external camera)
cap = cv2.VideoCapture(0)  # Change to 1 if you have multiple cameras

while cap.isOpened():
    ret, frame = cap.read()  # Capture a frame from the webcam
    if not ret:
        break  # Exit if the frame is not captured properly

    # Perform inference on the frame
    results = model(frame)

    # Display results
    annotated_frame = results[0].plot()  # Plot bounding boxes on frame

    # Show the output
    cv2.imshow("Fire Detection", annotated_frame)

    # Press 'q' to exit the camera feed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Input video path
input_video_path = "/content/4440816-hd_1920_1080_25fps.mp4"

# Output video path
output_video_path = "output_video_heatmap.mp4"

# Read the input video
cap = cv2.VideoCapture(input_video_path)

# Check if the video was opened successfully
if not cap.isOpened():
    print("Error: Could not open the video file.")
    exit()

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4 format

# Initialize the VideoWriter object to save the output video
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

print("Processing video...")

# Process each frame
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if no more frames

    # Apply heatmap to the frame
    heatmap_frame = cv2.applyColorMap(frame, cv2.COLORMAP_JET)

    # Write the processed frame to the output video
    out.write(heatmap_frame)

# Release resources
cap.release()
out.release()
print(f"Heatmap video saved to {output_video_path}")
```
